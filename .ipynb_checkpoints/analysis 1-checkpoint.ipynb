```python
ax = sns.countplot(data = df, x = 'Wkts')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Ov')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
    
    
#it decribes shows that mostly players has delivered 4 overs.
```

```python
ax=sns.countplot(data=df,x='Econ')
sns.set(rc={'figure.figsize':(30,20)})
for bars in ax.containers:
        ax.bar_label(bars)
```

```python
ax=sns.countplot(x="Runs",data=df)
```

```python
most_runs = df.groupby(['Runs'], as_index=False)['Wkts'].sum().sort_values(by='Wkts', ascending=False)

sns.barplot(x = 'Runs',y= 'Wkts' ,data = most_runs)
```

```python
most_runs = df.groupby(['Econ'], as_index=False)['Wkts'].sum().sort_values(by='Wkts', ascending=False)

sns.barplot(x = 'Econ',y= 'Wkts' ,data = most_runs)
```

```python
most_wickets = df.groupby(['Player'], as_index=False)['Wkts'].sum().sort_values(by='Wkts', ascending=False).head(10)

sns.barplot(x = 'Player',y= 'Wkts' ,data = most_wickets)
```

```python
most_wickets = df.groupby(['Player'], as_index=False)['Wkts'].sum().sort_values(by='Wkts', ascending=False).tail(10)

sns.barplot(x = 'Player',y= 'Wkts' ,data = most_wickets)
```

```python
most_wickets = df.groupby(['Venue'], as_index=False)['Wkts'].sum().sort_values(by='Wkts', ascending=False)

sns.barplot(x = 'Venue',y= 'Wkts' ,data = most_wickets)
```

```python
most_runs_by_venue = df.groupby(['Venue'], as_index=False)['Runs'].sum().sort_values(by='Runs', ascending=False)

sns.barplot(x = 'Venue',y= 'Runs' ,data = most_runs_by_venue)
```

```python
most_wickets = df.groupby(['Against'], as_index=False)['SR'].sum().sort_values(by='SR', ascending=False)

sns.barplot(x = 'Against',y= 'SR' ,data = most_wickets)
```
